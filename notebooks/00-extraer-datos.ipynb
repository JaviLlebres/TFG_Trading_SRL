```python
import yfinance as yf
import os
import pandas as pd
from datetime import datetime, timedelta

def preparar_datos_sp500():
    # 1. Crear la estructura de carpetas
    path_sp500 = "../data/SP500"
    if not os.path.exists(path_sp500):
        os.makedirs(path_sp500)
        print(f"✅ Carpeta creada: {path_sp500}")

    # 2. Configurar fechas (Máximo 729 días para datos de 1h)
    fecha_fin = datetime.now()
    fecha_inicio = fecha_fin - timedelta(days=729)
    
    start_str = fecha_inicio.strftime('%Y-%m-%d')
    end_str = fecha_fin.strftime('%Y-%m-%d')

    print(f"⏳ Descargando datos del S&P 500 (SPY) desde {start_str} hasta {end_str}...")

    # 3. Descarga de datos
    # Usamos interval='1h' y el rango calculado
    df = yf.download("SPY", start=start_str, end=end_str, interval="1h")

    if df.empty:
        print("❌ Error: No se descargaron datos. Verifica tu conexión a internet.")
        return

    # 4. Formatear para que sea IDÉNTICO a tus CSV de Cripto
    # Seleccionamos columnas y pasamos a minúsculas
    df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
    df.columns = ['open', 'high', 'low', 'close', 'volume']
    
    # Resetear índice para convertir el 'Datetime' en una columna
    df = df.reset_index()
    
    # Yahoo v0.2+ a veces llama a la columna 'Datetime' o 'index'
    # Renombramos la primera columna sea cual sea su nombre a 'timestamp'
    df.rename(columns={df.columns[0]: 'date'}, inplace=True)
    df['date'] = pd.to_datetime(df['date']).dt.tz_localize(None)
    # 5. Guardar el archivo
    # Para que sea más sencillo, el rango de fechas va a ser el mismo que el de bitcoin
    #BTCUSDT_1h_data_from_20210101_to_20251031
    nombre_archivo = f"SPY_1h_data_from_20210101_to_20251031.csv"
    full_path = os.path.join(path_sp500, nombre_archivo)
    
    df.to_csv(full_path, index=False)
    
    print("-" * 50)
    print(f"🚀 ¡Éxito! Datos guardados en: {full_path}")
    print(f"📊 Total de velas descargadas: {len(df)}")
    print(f"📅 Rango: {df['date'].min()}  >>>  {df['date'].max()}")
    print("-" * 50)

if __name__ == "__main__":
    preparar_datos_sp500()
```
